The following problem is solved using the Newton's Method.

The concentration of pollutant bacteria C in a lake decreases according to the given function. Determine the time required
for the bacteria concentration to be reduced to 5 using Newton's method.

c  =  70e^(-1.5t) + 25e^(-0.075t)

```python
import math

def f(x):
    # Enter formula
    fx = 70*math.e**(-1.5*x) + 25*math.e**(-0.075*x) - 5
    return fx

def fp(x):
    fpx = -105*math.e**(-1.5*x) - 1.875*math.e**(-0.075*x)
    return fpx

def normal_round(num, ndigits=0):
    return round(num + (10**(-1*(ndigits+2))), ndigits)
    
def start():
    prec = int(input("Enter number  of decimal places: "))
    x0 = normal_round(float(input("Enter value for x0: ")), prec)
    e = normal_round(float(input("Enter allowable error: ")), prec)
    
    # For restricting number of decimal places
    fstring = "{:." + str(prec) + "f}"
    
    fx0 = normal_round(f(x0), prec)
    fpx0 = normal_round(fp(x0), prec)
    
    lfx0 = []
    lpfx0 = []
    lx0 = []
    lx1 = []
    lx1x0 = []
    
    lx0.append(x0)
    
    print("\nf(x0) = ", fx0)
    print("f'(x0) = ", fpx0)
    
    header = ["Iteration", "f(x0)", "f'(x0)", "x0", "x1", "|x1-x0|"]
    lstring = "{:<15}" * (len(header))
    fstring = "{:." + str(prec) + "f}"
    
    n = 0
    while True:
        print("\nITERATION " + str(n))
        fx0 = normal_round(f(x0), prec)
        print("f(x0)  =  " + fstring.format(f(x0)))
        fpx0 = normal_round(fp(x0), prec)
        print("f'(x0)  =  " + fstring.format(f(x0)))
            
        lfx0.append(fx0)
        lpfx0.append(fpx0)
            
        x1 = normal_round(x0 - fx0/fpx0, prec)
        lx1.append(x1)
        print("x1  =  (" + fstring.format(x0) + " - (" + fstring.format(fx0) + " / "
             + fstring.format(fpx0) + ")  =  " + fstring.format(x1))

        print("x0 = ", fstring.format(x0), "     x1 = ", fstring.format(x1))
           
        lx1x0.append(abs(x1-x0))
        if (abs(x1 - x0) <= e):
            break
            
        x0 = x1
        lx0.append(x0)
        print("x1 -------> x0")
        print("")
        n += 1
            
    print("\nTabulated Results")
    print(lstring.format(*header))
    for i in range(len(lfx0)):
        print('%-15s%-15s%-15s%-15s%-15s%-15s' % (str(i), fstring.format(lfx0[i]), fstring.format(lpfx0[i]), 
                                                  fstring.format(lx0[i]), fstring.format(lx1[i]),
                                                 fstring.format(lx1x0[i])))
    print("\nThe  root is  equal to ", round(x1, prec))
        
start()
```
