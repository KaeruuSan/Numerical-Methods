```python
x = [5, 10, 15, 20, 25, 30]
fx_list = [[9, 13, 14, 10.5, 12, 13],
      [1.230229, 0.919948, 0.837915, 0.828062, 1.122035, 0.328179]]

prec = int(input("Enter number of decimal places: "))
fstring = "{:." + str(prec) + "f}"

data_pts = len(x)

def f(x):
    import math
    # Enter formula
    fx = x[0]*math.cos(x[1])
    print("\nF = f(x)cos[\u03B8(x)]")
    return fx

def normal_round(num, ndigits=0):
    return round(num + (10**(-1*(ndigits+2))), ndigits)

def m_poly(a, b):
    len_a = len(a)
    len_b = len(b)
    
    prod = []
    len_prod = len_a + len_b - 1
    
    for i in range(len_prod):
        prod.append(0)
        
    for i in range(len_a):
        for j in range(len_b):
            prod[i+j] += a[i] * b[j]
    return prod

def Lagrange(x, fx, term):
    numerator_calculated = []
    denominator_calculated = []
    
    print("----------TERM " + str(term) + "----------")
    print(str(data_pts) + " Data Points")
    print("f" + str(data_pts-1) + "(x)  =  ")
    for i in range(data_pts):
        numerator_str = ""
        denominator_str = ""
        numerator = []
        denominator = 1
        for j in range(data_pts):
            if i != j:
                numerator.append([1, -x[j]])
                numerator_str += "(x - " + str(x[j]) + ")"

                denominator *= (x[i] - x[j])
                denominator_str += "(" + str(x[i]) + " - " + str(x[j]) + ")"
        print("       [" + numerator_str + " / " + denominator_str + "]" + " * (" + str(fx[i]) +")")
        if i < data_pts - 1:
            print("    +")

        product = m_poly(numerator[0], numerator[1])
        for i in range(2, len(numerator)):
            product = m_poly(product, numerator[i])

        numerator_calculated.append(product)
        denominator_calculated.append(denominator)

    print("")
    total = []
    for i in range(len(numerator_calculated)):
        total.append(0)

    for i in range(len(numerator_calculated)):
        new_mult = normal_round(fx[i]/denominator_calculated[i], prec)
        for j in range(len(numerator_calculated[i])):
            total[j] += normal_round(numerator_calculated[i][j] * new_mult, prec)

    t_str = "f" + str(data_pts-1) + "(x)" + str(term) + "  =  "
    for i in range(len(total)):
        total[i] = normal_round(total[i], prec)
        if len(total) - i - 1 > 0:
            t_str += fstring.format(total[i]) + "x^" + str(len(total) - i - 1) + "  +  "
        else:
            t_str += fstring.format(total[i])
    print(t_str) 
    print("\n")
    
    return total

fit = []
for i in range(len(fx_list)):
    fit.append(Lagrange(x, fx_list[i], i+1))

new_x = float(input("\nEnter a new value for x: "))
new_y_list = []
z = 1
for j in fit:
    new_y = 0
    for i in range(len(j)):
        if i < len(j) - 1:
            new_y += j[i] * (new_x**(len(j) - (i+1)))
        else:
            new_y += j[i]
        new_y = normal_round(new_y, prec)
    new_y_list.append(new_y)
    print("f" + str(data_pts-1) + "(" + str(new_x)  + ")" + str(z) + "  =  " + fstring.format(new_y, prec))
    z += 1

print("  = " + fstring.format(f(new_y_list), prec))
```
