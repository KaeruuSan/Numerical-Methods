#### The following problem is solved using the Simpson's 3/8 Rule

Estimate the probability distribution on problem (1) uing the Multiple application Simpson's 3/8 Rule for the following segments n = 6, 9, 12 and 15 segments. Show the step-by-step solution and tabulate the results.
